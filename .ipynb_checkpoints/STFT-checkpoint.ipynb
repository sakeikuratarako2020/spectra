```python
%matplotlib inline

#L_neck_emg
target_ch = ["C3","C4","L_neck_emg","R_neck_emg"]
eeg_df = pd.DataFrame()
print(raw_fdt.get_data().shape)
raw_fdt_data = raw_fdt.get_data()
print(raw_fdt_data)
for ch in target_ch:
    raw_eeg = raw_fdt[ch,:]
    if not 'Time' in eeg_df.columns:
        eeg_df = pd.concat([eeg_df, pd.DataFrame({"Time":raw_eeg[1][:]})],axis=1)
    eeg_df = pd.concat([eeg_df,pd.DataFrame({ch:raw_eeg[0][0]})],axis=1)
    #print(raw_eeg[0][0])
    #print(raw_eeg[1][:])
    #plt.plot(raw_eeg[1][:])
    #plt.plot(raw_eeg[1][:],raw_eeg[0][0])
    
    display(eeg_df)
eeg_df.to_csv("eeg_df.csv")
```
